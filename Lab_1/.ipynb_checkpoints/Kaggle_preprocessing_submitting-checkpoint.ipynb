```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
data.info()
```

```python
def str_to_float(obj):
    global NUM_ERR

    try:
        return float(obj)
    except ValueError:
        NUM_ERR += 1

        try:
            return float(obj.replace(',', '.').replace('o', '0').replace('l', '1'))
        except ValueError:
            return float(obj.replace(' hundred', '')) * 100


def str_to_int(obj):
    global NUM_ERR

    try:
        return int(obj)
    except ValueError:
        NUM_ERR += 1

        return int(obj.replace('o', '0'))
```

```python
NUM_ERR = 0
```

```python
for col in ['Total day minutes', 'Total day calls', 'Total night minutes', 'Total intl minutes', 'Total charge']:
    data[col] = data[col].apply(str_to_float)
    print(NUM_ERR, 'errors for "{}"'.format(col))
    NUM_ERR = 0
```

```python
data['Customer service calls'] = data['Customer service calls'].apply(
    str_to_int)
print(NUM_ERR, 'errors for "{}"'.format('Customer service calls'))
NUM_ERR = 0
```

```python
data.info()
```

```python
from sklearn.impute import SimpleImputer
import numpy as np

imp_mode = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
data['Plans'] = imp_mode.fit_transform(np.array(data['Plans']).reshape(-1, 1))

imp_mode = SimpleImputer(missing_values=np.nan, strategy='mean')
data['Age'] = imp_mode.fit_transform(np.array(data['Age']).reshape(-1, 1))
```

```python
data['Area code'] = data['Phone number'].apply(lambda obj: obj[:3])
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
data['Sex'] = le.fit_transform(data['Sex'])
```
